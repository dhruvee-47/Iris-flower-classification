<h1><center><font color="#009999">Iris Flower Classification Using Decision Tree Algorithm : </font></h1></center>

- The goal of the project is to classify different species of iris flowers based on their petal and sepal dimensions.
- The dataset contains 150 samples of iris flowers, where each sample consists of four features - sepal length, sepal width, petal length, and petal width. The target variable is the species of iris flower, which can take three possible values - setosa, versicolor, and virginica.
- The project involves building a machine learning model that can accurately classify new samples of iris flowers based on their features. This can be done using various classification algorithms such as logistic regression, decision trees, k-nearest neighbors, support vector machines, or neural networks. 
- Here we are classifying using Decision Tree Algorithm. 
- The Iris Flowers dataset contains 150 samples of iris flowers, where each sample contains four features and a target variable indicating the species of the flower. The details of the variables in the dataset are as follows:
<ol>
    <li>Sepal Length (in cm): This variable represents the length of the sepal of the iris flower.</li>
    <li>Sepal Width (in cm): This variable represents the width of the sepal of the iris flower.</li>
    <li>Petal Length (in cm): This variable represents the length of the petal of the iris flower.</li>
    <li>Petal Width (in cm): This variable represents the width of the petal of the iris flower.</li>
    <li>Species: This variable represents the species of the iris flower. It can take one of three possible values         : setosa, versicolor, or virginica.</li>
</ol>
- The target variable, "Species," is a categorical variable, represented as a string.

<h2><center><font color="#730099">Spliting Dataset:</font></h2></center>

```python
x=df.drop('Species',axis=1)
y=df['Species']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.3,random_state=42)
```

<h2><center><font color="#730099">Importing DecisionTreeClassifier :</font></h2></center>

```python
from sklearn.tree import DecisionTreeClassifier 
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(xtrain,ytrain)
ypred=dtc.predict(xtest)
```

<h2><center><font color="#730099">Model Evaluation:</font></h2></center>

```python
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix
```

```python
print(f'Accuracy Score:= {accuracy_score(ytest,ypred)}')
```

```python
print("Classification_Report\n",classification_report(ypred,ytest))
```

<p><center><h3><font color="#ff9900">Checking for the overfitting & Underfitting : </h3></font></p></center>

```python
train= dtc.score(xtrain,ytrain)
test = dtc.score(xtest,ytest)
print(f"Training Accuracy:-{train}\nTesting Accuracy:-{test}")
```

<h3>- perfect model </h3>

<p><center><h1><font color="#730099">Scaling:</font></h1></p></center>

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler=StandardScaler()
xs_train = scaler.fit_transform(xtrain)
xs_test = scaler.transform(xtest)
```

```python
dtcs = DecisionTreeClassifier()
dtcs.fit(xs_train,ytrain)
ypred=dtcs.predict(xs_test)
print(f'Accuracy Score:- {accuracy_score(ytest,ypred)}')
print("Classification_report\n",classification_report(ypred,ytest))
```

```python
train=dtcs.score(xs_train,ytrain)
test = dtcs.score(xs_test,ytest)
print(f"Training Accuracy:-{train}\n Testing Accuracy:-{test}")
```

<p><center><h1><font color="#730099">HyperParameter Tunning:</font></h1></p></center>

```python
from sklearn.model_selection import GridSearchCV
```

```python
dtc=DecisionTreeClassifier()
param_grid = {
    "criterion":["gini","entropy"],
    "max_depth":list(range(1,20)),
    "min_samples_leaf":list(range(1,20))
}
grid=GridSearchCV(estimator=dtc,param_grid=param_grid,verbose=2)
grid.fit(xs_train,ytrain)
ypred=grid.predict(xs_test)
```

```python
grid.best_params_
```

```python
print(f'Accuracy Score:- {accuracy_score(ytest,ypred)}')
train = grid.score(xs_train,ytrain)
test = grid.score(xs_test,ytest)
print(f'Training Accuracy:-{train}\n Testing Accuracy:- {test}')
print()
print("Classification_report\n",classification_report(ypred,ytest))
```

```python
def predict_iris():
    sepl=float(input("Enter Sepal Length (cm)of Flower:- "))
    sepw=float(input("Enter Sepal Width (cm)of Flower:- "))
    petl=float(input("Enter Petal Length (cm)of Flower:- "))
    petw=float(input("Enter Petal Width (cm)of Flower:- "))
    newob = [[sepl,sepw,petl,petw]]
    yp= Final_model.predict(newob)
    print()
    if yp==0:
        print('Iris-setosa Flower')
    elif yp==1:
        print('Iris-versicolor Flower')
    else:
        print('Iris-virginica Flower')
    return yp
```
